# Model Training
Train LSTM/Transformer forecasters on synthetic surface sequences.

```python
import numpy as np
from src.datasets.surface_dataset import DatasetConfig, create_dataloaders
from src.models.base import ModelConfig
from src.models.factory import build_model
from src.training.trainer import Trainer, TrainConfig

surfaces = np.random.rand(220, 24, 41).astype('float32') * 0.3 + 0.1
train_loader, val_loader, test_loader = create_dataloaders(surfaces, DatasetConfig(lookback=20, horizon=1, batch_size=16))
model = build_model('transformer', ModelConfig(lookback=20, expiry_dim=24, strike_dim=41))
trainer = Trainer(model, train_loader, val_loader, TrainConfig(epochs=2))
fit = trainer.fit()
metrics = trainer.evaluate(test_loader)
fit['best_val_loss'], metrics
```
