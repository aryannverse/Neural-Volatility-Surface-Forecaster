# Surface Interpolation and Smoothing
Compare cubic spline and RBF interpolation.

```python
import numpy as np
from src.iv_surface.interpolation import SurfaceInterpolator, InterpolationConfig

x = np.linspace(0.8, 1.2, 20)
y = np.linspace(0.05, 1.2, 16)
z = np.sin(np.outer(y, x * 2.5)) * 0.03 + 0.2
z[2:5, 4:8] = np.nan

interp = SurfaceInterpolator(InterpolationConfig(method='cubic_spline'))
z_out = interp.interpolate_surface(x, y, z)
z_out.shape, np.isfinite(z_out).all()
```
