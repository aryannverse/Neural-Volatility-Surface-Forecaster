# Forecast Visualization
Interactive 3D and heatmap views for predicted vs actual surfaces.

```python
import numpy as np
from src.visualization.plotters import plot_surface_3d, plot_surface_heatmap

strike = np.linspace(0.8, 1.2, 41)
expiry = np.linspace(0.05, 1.5, 24)
actual = np.random.rand(24, 41) * 0.12 + 0.16
pred = actual + np.random.normal(0, 0.01, size=actual.shape)

fig_actual = plot_surface_3d(strike, expiry, actual, 'Actual Surface')
fig_pred = plot_surface_heatmap(strike, expiry, pred, 'Predicted Surface')
fig_actual, fig_pred
```
